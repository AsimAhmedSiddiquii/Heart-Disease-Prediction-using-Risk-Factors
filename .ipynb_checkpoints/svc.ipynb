```python
import numpy as np
import pandas as pd
from pandas import Series ,DataFrame
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn import metrics
import pickle
```

```python
dataset = pd.read_csv(r'dataset.csv')
```

```python
dataset
```

```python
dataset.shape
```

```python
train, test = train_test_split(dataset, test_size=0.3)
```

```python
print(train.shape)
```

```python
print(test.shape)
```

```python
train_x = train[['Sex','Age', 'Cholestrol','Pressure','Hereditary', 'Smoking', 'Alcohol', 'Physical', 'Diabetes', 'Diet', 'Obesity', 'Stress']]
train_y = train.HD
```

```python
test_x = train[['Sex','Age', 'Cholestrol','Pressure','Hereditary', 'Smoking', 'Alcohol', 'Physical', 'Diabetes', 'Diet', 'Obesity', 'Stress']]
test_y = train.HD
```

```python
train_x.head()
```

```python
train_y.head()
```

```python
model = svm.SVC()
model.fit(train_x,train_y)
prediction = model.predict(test_x)
```

```python
print(metrics.accuracy_score(prediction,test_y))
```
